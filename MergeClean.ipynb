```python
'''
ok so we will remove duplicates
      convert title and transcript to lowercase
      remove special character
      and yeah i think we will create a file and if exist we will skip
'''
import isodate
import pandas as pd
```

## Adding all the paths

```python
My_File_Path= '../Datasets/freecodecamp.csv'
Main_Data= '../Datasets/MainDataset.csv'
Output_Path='../Datasets/Cleaned.csv'
```

## Reading Files

```python
My_File =pd.read_csv(My_File_Path)
Main_F= pd.read_csv(Main_Data)
```

## Remove Duplicates

```python
#remove duplicates
Clean= My_File.drop_duplicates(subset='id')    
```

## Remove Special Characters and Convert to lowercase

```python
Clean['title']=(Clean['title'].str.lower().str.replace(r'[^a-z0-9 ]','',regex=True))
Clean['transcript']=(Clean['transcript'].str.lower().str.replace(r'[^a-z0-9 ]','',regex=True))
```

```python
print(Clean['duration'].unique())
```

```python
print(Main_F['channel_country'].unique())
```

## Good the list must be same length of column

```python
s_duration= []
for t in Clean['duration']:
    t = isodate.parse_duration(t).total_seconds()
    s_duration.append(t)
Clean['duration'] = s_duration
```

## adding extra columns of main dataset

```python
Clean['is_transcript_available']=(Clean['transcript']!=None)
Clean['duration_seconds']=Clean['duration']
```

```python
Clean.to_csv(Output_Path,index=False)
```

## Ok so time  for merge we use concat for stacking up the rows


```python
Clean = Clean[Main_F.columns]
```

```python
channel_id = Clean['channel_id'].iloc[0]
if channel_id not in Main_F['channel_id']: 
 Main_F= pd.concat([Main_F,Clean],ignore_index=True)
else:
    print("Channel already exist in the dataset")
```

```python
Main_F= pd.concat([Main_F,Clean],ignore_index=True)
```
